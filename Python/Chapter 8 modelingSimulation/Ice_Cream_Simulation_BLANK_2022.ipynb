#### Copy and paste code into a function
* Return true/false whether chocolate was chosen at least twice
* Abstraction!

```python
# Create function to simulate 5 customers
#Import the numpy package
import numpy as np

def singleCustomerTrial():
    # Create variables
    people = rand.randint(1,101)
    # For loop
    if (people <= 38):
        return "C"
    elif (people <= 80):
        return "V"
    else:
        return "S"

def simulate_5_customers():
    people = rand.randint(1,101)
    cNum = 0 
    vNum = 0
    sNum = 0
    for i in range(5):
        result = singleCustomerTrial()
        if result == "C":
            cNum += 1
        if result == "V":
            vNum += 1
        if result == "S":
            sNum += 1
    if(cNum > 1):
        return True
    return False

def simulate_n_customers(n=5, cMax=2):
    # Create variables
    people = rand.randint(1,101,n)
    cNum = 0 
    vNum = 0
    sNum = 0

    # For loop
    for customer in people:
        if (customer <= 38):
            cNum += 1
        elif (customer <= 80):
            vNum += 1
        else:
            sNum += 1

            
    # Print the end result of ice cream choices
    # print(f"There are {cNum} chocolates, {vNum} vanillas, and {sNum} strawberries.")

    # Print whether chocolate was chosen at least twice
    if(cNum >= cMax):
        return True
    return False

simulate_5_customers()
```

### Finding the probabilty that if 5 customers come in, at least two of them purchase chocolate
* Create an array to hold your results
* Create a loop to run 10000 times
    * Add result of each 5-customer trial to the array

```python
# Create variables
results = []


# Create a loop that repeats a large number of trials (5 customers), in this case 10000 times
for i in range(10000):
    results.append(simulate_n_customers())
# Print the results
print(results)
```

### Analyze the results

#### Find the total number of successes (true)

```python
# Create variables
successes = 0
failures = 0

# Loop through each entry, updating the number of successful trials (true)
for result in results:
    if result == True:
        successes += 1
    else:
        failures += 1
        
# Print the results
print(f"{successes} Successes, {failures} Fails")
```

#### Calculate the probability
* Probability = (successes/total # trials)

```python
import math

# Calculate probability
prob = successes/(successes+failures)

# Print results
print(f"{math.floor(prob*100)}%")
```

### Graph the results

```python
# Import the library
import matplotlib.pyplot as plot
# Magic to allow the graph to display directly in this notebook
%matplotlib inline

# Create an array of labels
labels = ["2 or more Chocolates", "Other"]

# Create an array of your results
result_data = [successes, failures]


# Explode option
    # 'Slices' appear distanced from the center
        # Larger numbers = further explosion
    # Explode array should be same size as labels
explode = (.05, 0)

# Use matplotlib module subplots() to get data for various charts
    # Returns a tuple in the form (figure,axes)
fig1,ax1 = plot.subplots()

# Use axes to create a pie chart
    # ax1.pie(data array, explode array, labels array, starting angle)
ax1.pie(result_data, explode, labels, autopct='%1.1f%%', startangle=90, colors=["darkgreen", "pink"], textprops={"color": "white", "size": "x-large"})


# Equal aspect ratio ensures that pie is drawn as a circle.
ax1.axis('equal')

plot.title("Probablity of 2 or more people out of 5 ordering Chocolate Icecream")

plot.show()
```
